```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
    
data = pd.read_csv("data/Systems.csv")
data = data[["Region",'Security']]

fig=plt.figure(figsize=(12,24), dpi= 100, facecolor='w', edgecolor='k')
plt.scatter(data["Security"],data["Region"], color="red", alpha = 0.5)
plt.xlabel("Security")
plt.ylabel("Region")
plt.show()
```
